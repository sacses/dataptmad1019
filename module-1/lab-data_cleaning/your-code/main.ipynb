#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
new_merged_users_posts.describe()
#We should ignorepostId and userId features in the DF, values they show ID numbers.

stats_merged = new_merged_users_posts.describe().T
stats_merged['MQC'] = stats_merged['75%'] - stats_merged['25%']
stats_merged
```

```python
outliers = pd.DataFrame(columns=new_merged_users_posts.columns)

for i in stats_merged.index:
    iqr = stats_merged.at[i,'MQC']
    cutoff = iqr * 1.5
    lower = stats_merged.at[i,'25%']
    upper = stats_merged.at[i,'75%']
    results = new_merged_users_posts[(new_merged_users_posts[i] < lower)|
                                    (new_merged_users_posts[i] > upper)].copy()
    results['Outlier'] = i
    outliers = outliers.append(results).reset_index(drop=True)
    
outliers

#Outcome is not correct, as features postId and userId should be not taken into consideration
```
